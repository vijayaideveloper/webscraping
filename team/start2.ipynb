```python
import pandas as pd
import sqlite3
from bs4 import BeautifulSoup
import requests
import time

data = []

for i in range(1,7):
    url = f"https://www.scrapethissite.com/pages/forms/?page_num={i}&per_page=100"
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")
    except Exception as e:
        print(f"Error Loading {e}")

    tables = soup.find("table", class_="table")
    table = tables.find_all("tr", class_="team")

    for each_table in table:
        team_name = each_table.find("td", class_="name").text.strip()
        year = each_table.find("td", class_="year").text.strip()
        wins = each_table.find("td", class_="wins").text.strip()
        losses = each_table.find("td", class_="losses").text.strip()
        ot_loss = each_table.find("td", class_="ot-losses").text.strip()
        pct = each_table.find("td", class_="pct").text.strip()
        gf = each_table.find("td", class_="gf").text.strip()
        ga = each_table.find("td", class_="ga").text.strip()
        diff = each_table.find("td", class_="diff").text.strip()
        data.append({
            "Team Name":team_name,
            "Year":int(year),
            "Wins":int(wins),
            "losses":int(losses),
            "Win %":float(pct),
            "OT Losses":ot_loss,
            "Goals For (GF)":int(gf),
            "Goals Against (GA)":int(ga),
            "Differrence":int(diff)
        })
        
    print(f"Page {i} is Completed ‚úÖ")
    time.sleep(1)
df = pd.DataFrame(data)

df.to_csv("Teams_report.csv", index=False)

df.to_excel("Team_report.xlsx", index=False)

df.to_json("Team_report.json", orient="records", indent=4)

connection = sqlite3.connect("team.db")
df.to_sql("team_report", connection, if_exists="replace", index=False)
connection.close()

print("Successfully Saved üëç")
```
