# A* Memory Bounded Algorithm
## Types:
### IDA --> Iterative Depening A* 
It used DFS
### RDFS --> Recursive Best First Search
It uses Recursive approch
### SMA* --> Simplified Memory Bound A*
It reduceses memory usage by limitting memory

```python
import heapq
def heuristic(n,goal):
    return abs(n-goal)

def astar(graph,start,goal):
    open_list = []
    heapq.heappush(open_list,(0,start))
    g = {start:0}
    parent = {start:None}

    while open_list:
        f,current = heapq.heappop(open_list)

        if current == goal:
            path = []

            while current:
                path.append(current)
                current = parent[current]
            return path[::-1]

        for neighbor,cost in graph[current]:
            new_g = g[current] + cost
            
            if neighbor not in g or new_g < g[neighbor]:
                g[neighbor] = new_g
                f = new_g + heuristic(neighbor,goal)
                heapq.heappush(open_list,(f,neighbor))
                parent[neighbor]=current
    return None

graph = {
    0:[(1,1), (2,4)],
    1:[(2,2), (3,5)],
    2:[(3,1)],
    3:[]
}

print("Path : ", astar(graph,0,3))
```

# Min Max Algorithm

<img src='minmax.png'>


It is a algorithm used in two player games
#### Maximizer -- > Tries to increase score
#### Minimizer -- > Tries to reduce score

```python
def minimax(depth, node_index, is_max, values, max_depth, alpha, beta):
    # Base case: leaf node reached
    if depth == max_depth:
        return values[node_index]

    if is_max:
        best = float('-inf')
        
        for i in range(2):
            val = minimax(depth + 1, node_index * 2 + i, False, values, max_depth, alpha, beta)
            best = max(best, val)
            alpha = max(alpha, best) # alpha --> best max value
            # Pruning
            if beta <= alpha:
                break
        return best
    
    else:
        best = float('inf') 
        for i in range(2):
            val = minimax(depth + 1, node_index * 2 + i, True, values, max_depth, alpha, beta)
            best = min(best, val)
            beta = min(beta, best) # beta --> best min value
            # Pruning
            if beta <= alpha:
                break
        return best

values = [4, 5, 3, 7, -1, 0, 8, 9]
# Call with alpha as -inf and beta as +inf
result = minimax(0, 0, True, values, 3, float('-inf'), float('inf'))

print('Optimal Value:', result)
```

# Alpha Beta Pruning
**Alpha Beta Pruning** is used to reduce **time and space complexity** by removing useless branchs
<img src='alpha_beta.webp' width='600' height='500'>

```python
def minimax(depth, node_index, is_max, values, max_depth, alpha, beta):
    # Base case: leaf node reached
    if depth == max_depth:
        return values[node_index]

    if is_max:
        best = float('-inf')
        # Check both left (index*2) and right (index*2 + 1) children
        for i in range(2):
            val = minimax(depth + 1, node_index * 2 + i, False, values, max_depth, alpha, beta)
            best = max(best, val)
            alpha = max(alpha, best) # alpha --> best max value
            # Pruning
            if beta <= alpha:
                break
        return best
    
    else:
        best = float('inf') # Corrected: Minimizer starts at +infinity
        # Check both left (index*2) and right (index*2 + 1) children
        for i in range(2):
            val = minimax(depth + 1, node_index * 2 + i, True, values, max_depth, alpha, beta)
            best = min(best, val)
            beta = min(beta, best) # beta --> best min value
            # Pruning
            if beta <= alpha:
                break
        return best

values = [4, 5, 3, 7, -1, 0, 8, 9]
# Call with alpha as -inf and beta as +inf
result = minimax(0, 0, True, values, 3, float('-inf'), float('inf'))

print('Optimal Value:', result)
```

```python
import tkinter as tk
from tkinter import messagebox
import random

# Create window
root = tk.Tk()
root.title("Tic-Tac-Toe AI")

board = [' ' for _ in range(9)]
buttons = []

# Check winner
def check_winner(b):
    win_conditions = [
        (0,1,2),(3,4,5),(6,7,8),
        (0,3,6),(1,4,7),(2,5,8),
        (0,4,8),(2,4,6)
    ]

    for i,j,k in win_conditions:
        if b[i] == b[j] == b[k] and b[i] != ' ':
            return b[i]
    return None

def is_full(b):
    return ' ' not in b

# Minimax
def minimax(b, is_max):
    winner = check_winner(b)

    if winner == 'O':
        return 1
    elif winner == 'X':
        return -1
    elif is_full(b):
        return 0

    if is_max:
        best = -100
        for i in range(9):
            if b[i] == ' ':
                b[i] = 'O'
                score = minimax(b, False)
                b[i] = ' '
                best = max(best, score)
        return best
    else:
        best = 100
        for i in range(9):
            if b[i] == ' ':
                b[i] = 'X'
                score = minimax(b, True)
                b[i] = ' '
                best = min(best, score)
        return best

# Best move for AI
import random

def best_move():
    # 50% chance to play random move (easy)
    if random.random() < 0.5:
        empty = [i for i in range(9) if board[i] == ' ']
        return random.choice(empty)

    # Otherwise use minimax (smart move)
    best_score = -100
    move = -1

    for i in range(9):
        if board[i] == ' ':
            board[i] = 'O'
            score = minimax(board, False)
            board[i] = ' '
            if score > best_score:
                best_score = score
                move = i

    return move

# Button click
def click(i):
    if board[i] == ' ':
        board[i] = 'X'
        buttons[i].config(text='X')

        if check_game():
            return

        ai = best_move()
        if ai != -1:
            board[ai] = 'O'
            buttons[ai].config(text='O')

        check_game()

# Check game status
def check_game():
    winner = check_winner(board)

    if winner:
        messagebox.showinfo("Game Over", f"{winner} wins!")
        reset()
        return True

    if is_full(board):
        messagebox.showinfo("Game Over", "Draw!")
        reset()
        return True

    return False

# Reset game
def reset():
    global board
    board = [' ' for _ in range(9)]
    for btn in buttons:
        btn.config(text=' ')

# Create buttons
for i in range(9):
    btn = tk.Button(root, text=' ', font=('Arial', 20), width=5, height=2,
                    command=lambda i=i: click(i))
    btn.grid(row=i//3, column=i%3)
    buttons.append(btn)

root.mainloop()
```
