```python
import pandas as pd
import requests
import time
import sqlite3
```

```python
start = 2010
end = 2015
data = []
for i in range(start, end+1):
    url = f"https://www.scrapethissite.com/pages/ajax-javascript/?ajax=true&year={i}"
    headers = {
        'User-Agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36'), 'Accept-Encoding': 'gzip, deflate', 'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8', 'Connection': 'keep-alive', 'Accept-Language': 'en-US,en;q=0.6'
    }


    response = requests.get(url, headers=headers)

    films = response.json()
    for film in films:
        title = film['title']
        year = film['year']
        awards = film['awards']
        nomination = film['nominations']
        data.append(
            {
                "Title":title,
                "Year":int(year),
                "Nomination":int(nomination),
                "Awards":int(awards)
            }
        )
    time.sleep(1)
```

```python
df = pd.DataFrame(data)
print(df.head(20))
```

```python
df.to_csv("Oscar_movies.csv", index=False)

df.to_excel("Oscar_movies.xlsx", index=False)

df.to_json("Oscar_movie.json", orient='records', indent=4)

connection = sqlite3.connect("Oscar_movie.db")
df.to_sql("Oscars Movies", connection, if_exists="replace", index=False)
connection.close()
```
