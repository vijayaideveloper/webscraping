```python
import pandas as pd
import sqlite3
from bs4 import BeautifulSoup
import requests
import time



url = "https://www.scrapethissite.com/pages/simple/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")


countrys = soup.find_all("div", class_="row")
countrys = countrys[3:-1]


data = []


for country in countrys:


    main = country.find_all("div", class_="country")
    for m in main:
        name = m.find("h3").text.strip()
        capital = m.find("div", class_="country-info").find("span", class_="country-capital").text.strip()
        population = m.find("div", class_="country-info").find("span", class_="country-population").text.strip()
        area = m.find("div", class_="country-info").find("span", class_="country-area").text.strip()
        print(f'{name} is Collected !')
        data.append({
        "Country":name,
        "Capital":capital,
        "Population":int(population),
        "Area":float(area)
    })
time.sleep(1)




df = pd.DataFrame(data)

df.to_csv("Country.csv", index=False)
print("Succesfully CSV Saved !")

df.to_excel("Country.xlsx", index=False)
print("Successfully excel Saved !")

df.to_json("Country.json", orient='records', indent=4)
print("Successfully json Saved !")


connection = sqlite3.connect("country.db")
df.to_sql("Countries_data", connection, if_exists="replace", index=False)
connection.close()
print("Successfully Sql Saved !")
```
