## Statistics for AI engineers

### Mean vs Median
Box Office mean = ₹137Cr, but median = ₹54Cr. The mean is dragged up by blockbusters like ₹1600Cr hits. For skewed data, median is more honest.

```python
print(data['Box Office'].mean()) # 136.6
data['Box Office'].median() # 53.65 ← use this
```

### Standard Deviation & Outliers
Std = 242. That's higher than the mean (137) — extreme spread. Outlier detection:

```python
Q1 = data['Box Office'].quantile(0.25)
Q1
```

```python
Q3 = data['Box Office'].quantile(0.75)
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
#outliers = df[df['Box Office'] > Q3 + 1.5*IQR]
outliers = data[data['Box Office'] > Q3 + 1.5 * IQR]
outliers
```

```python
Q3 + 1.5 * IQR
```

### Correlation
Does higher budget = higher box office? Correlation tells you.

### Class Imbalance
Verdict distribution: Flop=52, Hit=43, Average=16, Super Hit=16, Blockbuster=10. This is imbalanced — your ML model will be biased toward "Flop."

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np
```

```python
# Step 1 — get the unique classes and target column
classes = np.array(data['Verdict'].unique())
y = data['Verdict']
```

```python
# Step 2 — compute weights
weights = compute_class_weight('balanced',classes=classes,y=y)
```

```python
# Step 3 — make it a dictionary (this is what sklearn models expect)
class_weight_dict = dict(zip(classes, weights))
class_weight_dict
```

**What this means**: The model is told to treat one "All Time Blockbuster" example as if it were ~6 Flop examples. This forces the model to pay more attention to rare classes.

## Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Tamil Movies Dataset - EDA', fontsize=16, fontweight='bold')

# 1. Box Office histogram
axes[0,0].hist(data['Box Office'], bins=30, color='steelblue', edgecolor='white')
axes[0,0].set_title('Box Office Distribution')
axes[0,0].set_xlabel('Box Office (Cr)')
axes[0,0].set_ylabel('Count')

# 2. Verdict count bar chart
verdict_counts = data['Verdict'].value_counts()
axes[0,1].bar(verdict_counts.index, verdict_counts.values, color='coral', edgecolor='white')
axes[0,1].set_title('Verdict Distribution')
axes[0,1].set_xlabel('Verdict')
axes[0,1].set_ylabel('Count')
axes[0,1].tick_params(axis='x', rotation=45)

# 3. Budget vs Box Office scatter (colored by Verdict)
verdicts = data['Verdict'].unique()
colors = plt.cm.tab10(range(len(verdicts)))
for verdict, color in zip(verdicts, colors):
    subset = data[data['Verdict'] == verdict]
    axes[1,0].scatter(subset['Budget'], subset['Box Office'],
                      label=verdict, color=color, alpha=0.7)
axes[1,0].set_title('Budget vs Box Office')
axes[1,0].set_xlabel('Budget (Cr)')
axes[1,0].set_ylabel('Box Office (Cr)')
axes[1,0].legend(fontsize=7)

# 4. Average Box Office by Verdict
avg_bo = data.groupby('Verdict')['Box Office'].mean().sort_values()
axes[1,1].barh(avg_bo.index, avg_bo.values, color='mediumseagreen', edgecolor='white')
axes[1,1].set_title('Avg Box Office by Verdict')
axes[1,1].set_xlabel('Avg Box Office (Cr)')

plt.tight_layout()
plt.savefig('eda_plots.png', dpi=150, bbox_inches='tight')
plt.show()
```

1. **Top left** — Box Office histogram (you'll see the right skew clearly)
2. **Top right** — Verdict bar chart (class imbalance visible instantly)
3. **Bottom left** — Scatter plot of Budget vs Box Office, each Verdict in a different color
4. **Bottom right** — Horizontal bar of average Box Office per Verdict category
